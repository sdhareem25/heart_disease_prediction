EXPLORING REALTIONSHIP:HEATMAPS WITH PYTHON FOR DATA VISUALIZATION

```python
plt.figure(figsize=(16,6))
sns.heatmap(df.corr(),annot=True)
```

NUMBER AFFECTED AND UNAFFECTED BY HEART PROBLEMS

```python
sns.countplot(x='target', data=df, hue='target', palette=['yellow', 'red'], legend=False)
```

GENDER DISTRIBUTION IN DATASET

```python
sns.countplot(x='sex', data=df, hue='sex', palette=['pink','blue'], legend=False)
```

DISTRIBUTION OF HEAR DISEASE AMONG MALES AND FEMALE

```python
sns.countplot(data=df,x=df['sex'],hue=df['target'])
plt.xticks([0,1],['female','males'])
plt.legend(labels=['no-disease','disease'])
plt.show()
```

DISTRIBUTION OF AGES IN DATASET

```python
sns.histplot(df['age'], bins=20, kde=True)
plt.show()
```

CHEST PAIN TYPES

```python
sns.countplot(x='cp',data=df)
plt.xticks([0,1,2,3],['typical angina','atypical angina','non-anginal','asympatomatic'])
plt.show()
```

CHEST PAIN DISTRIBUTION IN HEART DISEASE VS NON-HEART DISEASE

```python
sns.countplot(data=df,x=df['cp'],hue=df['target'])
plt.xticks([0,1,2,3],['typical angina','atypical angina','non-anginal','asympatomatic'])
plt.legend(labels=['no heart disease','disease'])
plt.show()
```
